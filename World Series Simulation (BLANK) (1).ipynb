## Defining an individual trial
#### Copy and paste code into a function
* Return a tuple in the format (winning league, total number of games played)
* Abstraction!

```python
# Create function to represent one trial of a World Series 
def single_trial():
    import numpy as np
    AL_win_count = 0
    NL_win_count = 0
    gameCount = 0
    win = False
    # While loop

    while win == False:
        winnerOfGame = single_game_trial()
        if NL_win_count < 4 and winnerOfGame=="NL wins":
            NL_win_count += 1
            gameCount += 1
            
        elif AL_win_count < 4 and winnerOfGame =="AL wins":
            AL_win_count += 1
            gameCount += 1
            
        if AL_win_count==4:
            winner = "AL wins the series"
            win = True
            
        if NL_win_count==4:
            winner = "NL wins the series"
            win = True
    # Total games played before a division reached four wins

    # Print winner and games
    result = (winner, gameCount)
    return result
```
